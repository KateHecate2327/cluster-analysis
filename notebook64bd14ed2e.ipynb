```python
pay_bonus = mall_data['pay_bonus'] 

from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

#df = pd.DataFrame({" rating": [1, 2, 3, 4, 6, 7, 9, 9, 9, 10], "dummy": range(10)})
plt.figure(figsize=(10,5))
age_bins = range(10,1000,50)
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, figsize=(50,25), gridspec_kw= {"height_ratios": (5000, 10000)})

mean=mall_data['pay_bonus'].mean()
median=mall_data['pay_bonus'].median()
mode=mall_data['pay_bonus'].mode().values[0]

sns.boxplot(data=mall_data, x="pay_bonus", ax=ax_box)
ax_box.axvline(mean, color='r', linestyle='--')
ax_box.axvline(median, color='g', linestyle='-')
ax_box.axvline(mode, color='b', linestyle='-')

sns.histplot(data=mall_data, bins=age_bins, x="pay_bonus", ax=ax_hist, kde=False, binwidth=5, alpha=0.5)
ax_hist.axvline(mean, color='r', linestyle='--', label="Mean")
ax_hist.axvline(median, color='g', linestyle='-', label="Median")
ax_hist.axvline(mode, color='b', linestyle='-', label="Mode")
ax_hist.text(750,190, "Mean : {:.1f}".format(pay_bonus.mean()))
ax_hist.text(650,190, "Median : {:.1f}".format(pay_bonus.median()))
ax_hist.text(850,190, "Std. deviation: {:.1f}".format(pay_bonus.std()))
#ax_hist.text(950,190, "Mode: {:.1f}".format(pay_bonus.mode()))
ax_hist.legend()

ax_box.set(xlabel='pay_bonus')
plt.show()
```

```python
my_bins = range(10,3500,10)
pay_bonus = mall_data['pay_bonus'] 
cash_price = mall_data['cash_price'] 
# males histogram
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18,5))
sns.distplot(pay_bonus, bins=my_bins, kde=False, color='#0066ff', ax=ax1, hist_kws=dict(edgecolor="k", linewidth=2))
ax1.set_xticks(my_bins)
ax1.set_yticks(range(0,2400,2))
ax1.set_ylim(0,2200)
ax1.set_title('pay_bonus')
ax1.set_ylabel('Count')
ax1.text(85,19, "Mean : {:.1f}".format(pay_bonus.mean()))
ax1.text(85,18, "Median : {:.1f}".format(pay_bonus.median()))
ax1.text(85,17, "Std. deviation: {:.1f}".format(pay_bonus.std()))

# females histogram
sns.distplot(cash_price, bins=my_bins, kde=False, color='#cc66ff', ax=ax2, hist_kws=dict(edgecolor="k", linewidth=2))
ax2.set_xticks(my_bins)
ax2.set_yticks(range(0,2400,2))
ax2.set_ylim(0,2200)
ax2.set_title('cash_price')
ax2.set_ylabel('Count')
ax2.text(85,19, "Mean : {:.1f}".format(cash_price.mean()))
ax2.text(85,18, "Median : {:.1f}".format(cash_price.median()))
ax2.text(85,17, "Std. deviation: {:.1f}".format(cash_price.std()))

```
